```python
import json
import re
```

```python
with open('../../bundestagsredeN parser/speeches_20.jsonl', 'r', encoding='utf8') as fp:
    data = list(fp)
speeches = []
for line in data:
    speeches.append(json.loads(line))
```

```python
kommentare = []
for speech in speeches:
    #if speech['party'] == 'AfD':
        text = speech['text']
        matches = re.findall("\(([^\)]+)\)",text)
        kommentare.extend(matches)
#print(kommentare)
kommentare
```
